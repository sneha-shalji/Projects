This Analysis aims to serve as a benchmark for an e-commerce fashion store. Using this, you may want to try and understand what you can expect of your users and determine in advance how your grows may be.

# **Segmentation of users**

Currently, the dataset contains too many users who are inactive (users that only seem to have signed up to try the site out, maybe browse few articles and do nothing else/drop out). The website actually strongly incitate users to sign up (otherwise they cannot browse the site for long), which may be a reason for such a proportion of passive users.

We will be Dividing the users as **Active users** and **passive users**

Active users consist of:

*   Potential Buyers
*   Potential Sellers

From now on we will analyse Seller-specific and Buyer-Secific Behaviour

```python
# creating a dataframe called 'buyers' where Users have brought atleast 1 product
buyers= df[df.productsBought > 0]
buyers
```

```python
### Prospecting Sellers (has sold a product or is still trying do his/her first sale)
sellers= df[(df.productsListed > 0) | (df.productsSold > 0)]
sellers
```

```python
### Successful sellers (sold at least 3 products)
successful_sellers= df[df.productsSold >3]
successful_sellers
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
sns.countplot(x='productsWished',hue='productsWished',legend=False,data=buyers,palette='plasma')
plt.xlabel("No of products the user Wish Listed")
plt.subplot(1,2,2)
sns.countplot(x='productsBought',hue='productsBought',legend=False,data=buyers,palette='plasma')
plt.xlabel("No of products bought")
plt.suptitle("Relation betwen ProductsWished and ProductsBought")
```

```python
dropout_after = lambda df, dayMax: dayMax - df.daysSinceLastLogin
max_last_login = df.daysSinceLastLogin.max()
```

```python
plt.figure(figsize=(15,7))
sns.kdeplot(np.array(dropout_after(df, max_last_login)), fill=True)
plt.title(f"User retention : how users drop out over time")
plt.show()
```

At First we can see a lot of users were active but after about 100 days vast amount of them abandoned the site( Maybe because the products weren't much to their liking)
The spike at ~700 shows that among all the users who signed up during the same week/day, the vast majority of them dropped right away.
This may have happend because of some promotional event or sale. A slight increse in the activity can be seen. But they also dropped after some time.


```python
plt.figure(figsize=(15,7))
sns.kdeplot(np.array(dropout_after(buyers, max_last_login)), fill=True)
plt.title("User retention - how people who bought at least one product dropout")
plt.xlabel("(past) <=  # days before dropout  => (present)")
plt.ylabel("density")
plt.show()

plt.figure(figsize=(15,7))
sns.kdeplot(np.array(dropout_after(sellers, max_last_login)), fill=True)
plt.title("User retention - how sellers keep visiting the site")
plt.xlabel("(past) <=  # days before dropout  => (present)")
plt.ylabel("density")
plt.show()

plt.figure(figsize=(15,7))
sns.kdeplot(np.array(dropout_after(sellers,max_last_login)), fill=True)
sns.kdeplot(np.array(dropout_after(buyers,max_last_login)), fill=True)
plt.title("User retention - buyers and sellers: dropout curve")
plt.xlabel("(past) <=  # days before dropout  => (present)")
plt.ylabel("density")
plt.legend(["sellers", "buyers"])
plt.show()
```

We see that even a one-time buyer can be expected to keep visiting you over a long period of time

**sellers** : The platform is easy to use for your sellers/service providers
so they keep continue using the site. whenever there is a event or a sale they become extremly active on the website

**buyers**: mainly two distinct behavior: purchase once and then never come back (as denoted by the first litlle bump) or staying faithful to the platform for a long while (the spike at the end representing the present).

High Skewness is Exhibited by some of the columns.

This is mainly Because of the irregularity of the users' actions in the platform.

For instance, when a new user signs up for the site the site shows them the 3 most popular users to follow which the new users follows right away. But this makes 3 the standard value or the normal value in SocialNbFollows and All the other values outliers. But that is incorrect.


Thus the Outliers present in the Boxplot are not Outliers, but our crucial data(The active users in the platform).

### **Removing Outliers**

```python
sns.boxplot(df['socialNbFollows'])
```

```python
# I am using z-score method here because all the other methods completely removes the Outliers.
# Defining Limits
upper_limit=df['socialNbFollows'].mean() + 3*df['socialNbFollows'].std()
lower_limit=df['socialNbFollows'].mean() - 3*df['socialNbFollows'].std()
print("upper limit:",upper_limit)
print("lower limit:",lower_limit)
```

```python
# Finding Outliers
df.loc[(df['socialNbFollows']>upper_limit) | (df['socialNbFollows']<lower_limit)].reset_index(drop=True)
```

```python
# Removing the Outliers
df['socialNbFollows']=np.clip(df['socialNbFollows'],lower_limit,upper_limit)
```

# **Model Creation**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay,classification_report
knn_model=KNeighborsClassifier(n_neighbors=7,weights='distance')
log_model=LogisticRegression(multi_class='auto',solver='newton-cg')
Nb_model=GaussianNB()
dt_model=DecisionTreeClassifier(criterion='entropy')
rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
model_names=["KNeighborsClassifier","LogisticRegression","GaussianNB","DecisionTreeClassifier","RandomForestClassifier"]
a_models=[knn_model,log_model,Nb_model,dt_model,rf_model]
lst_score=[]
```

```python
for i in a_models:
  print("Model:",i)
  print('-'*150)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  mat=confusion_matrix(y_test,y_pred)
  score=accuracy_score(y_test,y_pred)
  lst_score.append(score)
  print("Accuracy score:",score)
  print("Model Evaluation Result:")
  print(classification_report(y_test,y_pred))
  print("="*150)
```

```python
lst_score
```

```python
models_df=pd.DataFrame({"Model":model_names,"Accuracy score":lst_score})
best_models=models_df.sort_values(by='Accuracy score',ascending=False)
best_models
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=best_models['Model'],y=best_models['Accuracy score'],hue=best_models['Accuracy score'],data=best_models,palette='tab10')
plt.xlabel("Models")
plt.ylabel("Accuracy Score")
```
